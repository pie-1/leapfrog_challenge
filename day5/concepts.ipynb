## lazy loading and pagination


- **Lazy Loading**  
  - Component haru suruma load nagari, user le use garne bela matra load garne technique ho.  
  - Yo le app lai fast ra lightweight banaucha.

- **React.lazy()**  
  - React ko built-in function ho jasko help le hamile component lai lazy load garna sakchau.  
  - Example: `const PokemonCard = lazy(() => import('./PokemonCard'))`

- **Suspense**  
  - Lazy component load hudai garda "Loading..." message dekhाउने wrapper ho.  
  - Without Suspense, lazy loading ma error aaucha.

- **Pagination**  
  - Dherai data lai page page ma divide garera dekhाउने system.  
  - Example: 12 Pokémon per page (Next & Previous button)

- **Context API**  
  - Component haru bich ma data share garne easy tarika.  
  - Prop drilling avoid garna madat garcha.

- **useContext Hook**  
  - Context API bata data tanne hook ho.  
  - App.jsx ma use garera context ko value use gareko.

- **useState + useEffect**  
  - `useState` → data store garna  
  - `useEffect` → page change huda naya data fetch garna

- **API Integration (Axios)**  
  - PokeAPI bata real Pokémon data tanne.  
  - `axios.get()` use garera backend jasto data lyaune.

- **Client-side Search Filtering**  
  - User le search garda list filter garne.  
  - `.filter()` ra `.includes()` use garera instant search.

- **Framer Motion**  
  - Card hover ma smooth animation dincha (scale, rotate).  
  - App lai alive ra engaging banaucha.

- **Responsive Grid Layout**  
  - Mobile, tablet ra desktop ma ramro dekhine grid system.  
  - `grid-cols-1 sm:grid-cols-2 lg:grid-cols-4`
